```python
df1 = df.drop(['Loan_ID','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_income','CoapplicantIncome_log'],axis=1)
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
columns = ['Gender','Dependents','Married','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
```

```python
for col in columns:
    df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop('Loan_Status',axis='columns')
y = df1.Loan_Status
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
def classify(model,x,y):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
    model.fit(x_train,y_train)
    print('Model Score = ', model.score(x_test,y_test)*100)
    
    score = cross_val_score(model,x,y,cv=5)
    print('Cross Validation score = ', score.mean()*100)
```

```python
from sklearn.linear_model import LogisticRegression
le = LogisticRegression()
classify(le,x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
classify(tree,x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
classify(rf,x,y)
```

# Hyper parameter tuning

```python
rf1 = RandomForestClassifier(n_estimators=100,min_samples_split=25)
classify(rf1,x,y)
```
